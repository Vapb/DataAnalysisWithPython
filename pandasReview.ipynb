```python
# Country Canada and Sub_Category is Bike Racks
df.loc[(df['Country'] == 'Canada') & (df['Sub_Category'] == 'Bike Racks')].shape

# Country France and State
df.loc[df['Country'] == 'France'].shape

# How many orders were made per accessory sub-categories?
df.loc[df['Product_Category'] == 'Accessories']['Sub_Category']
df.loc[df['Product_Category'] == 'Accessories', 'Sub_Category']

# How many sales with more than 500 in Revenue were made by men?
df.loc[(df['Customer_Gender'] == 'M') & (df['Revenue'] == 500)]

# Get the sale with the highest revenue
cond = df['Revenue'] == df['Revenue'].max()
df.loc[cond]

# How many orders were made in May of 2016?
cond = (df['Year'] == 2016) & (df['Month'] == 'May')
df.loc[cond].shape[0]

# How many orders were made between May and July of 2016?
cond = (df['Year'] == 2016) & (df['Month'].isin(['May', 'June', 'July']))
df.loc[cond].shape[0]

# Add 7.2% TAX on every sale Unit_Price within United States
df.loc[df['Country'] == 'United States', 'Unit_Price'] *= 1.072

# How many rentals were made at the store with id 2 or with replacement cost lower than 10.99 USD?
df.loc[(df['store_id'] == 2) | (df['film_replacement_cost'] < 10.99)].shape[0]
```
